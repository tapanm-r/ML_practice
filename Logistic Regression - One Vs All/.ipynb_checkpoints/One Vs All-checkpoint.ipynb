```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("OneVAll-IRIS.csv")

data.head()
```

```python
set(data['species'].to_numpy())
```

```python
species = {'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3}

data.species = [species[item] for item in data.species]

data.describe()
```

```python
m = len(data['species'])
n = len(data.columns) - 1

X = data.iloc[:, :-1].to_numpy().reshape((m, n))
y = data.iloc[:, -1].to_numpy().reshape((m, 1))
```

```python
for i in range(n):
    mean_x = np.mean(X[:][i])
    std_x = np.std(X[:][i])
    for j in range(m):
        X[j][i] = (X[j][i] - mean_x) / std_x
```

```python
ones = np.ones((m, 1))

X = np.concatenate((ones, X), axis = 1)

X.shape
```

```python
def train_test_split(X, y, s = 0.3):
    assert(X.shape[0] == y.shape[0])
    m = X.shape[0]
    val = int(m - m * s)
    
    indices = np.arange(m)
    np.random.shuffle(indices)

    A = X[indices]
    B = y[indices]
    
    X_train = A[:val + 1]
    y_train = B[:val + 1]
    
    X_test = A[val + 1:]
    y_test = B[val + 1:]
    
    return X_train, y_train, X_test, y_test
```

```python
X_train, y_train, X_test, y_test = train_test_split(X, y)
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))
```

```python
def hyp(theta, X):
    h = sigmoid(X @ theta)
    
    return h
```

```python
def cost(theta, X, y, m):
    J = -1 * np.sum(y.T @ np.log(hyp(theta, X)) + (1 - y).T @ np.log(1 - hyp(theta, X))) / m
    return J
```

```python
def train_model(theta, X, y, m, lr = 0.1):
    J = []
    cost0 = cost(theta, X, y, m)
    J.append(cost0)
    for i in range(10000):
        delta = X.T @ (hyp(theta, X) - y) / m
        theta = theta - lr * delta
        
        cost1 = cost(theta, X, y, m)
        J.append(cost1)
        if cost1 == 0 or cost1 > cost0:
            print(i)
            break
        cost0 = cost1
    
    return J, theta
```

```python
def one_vs_all(X, y):
    m = X.shape[0]
    n = len(set([item[0] for item in y]))
    thetas = np.ones((X.shape[1], n))
    
    J = []
    for i in range(n):
        h = np.zeros_like(y)
        h[y == i + 1] = 1.0
        theta = thetas[:, i].reshape((X.shape[1], 1))
        c, theta = train_model(theta, X, h, m)
        J.append(c)
        thetas[:, i] = np.ravel(theta)
    
    return J, thetas
```

```python
def pred(thetas, X):
    h = hyp(thetas, X)
    y = np.zeros((X.shape[0], 1))
    
    for i in range(X.shape[0]):
        y[i] = np.argmax(h[i]) + 1
    
    return y
```

```python
J, thetas = one_vs_all(X_train, y_train)

thetas
```

```python
y_pred = pred(thetas, X_test)
accuracy = 0
n = X_test.shape[0]

for i in range(n):
    if y_pred[i] == y_test[i]:
        accuracy += 1

accuracy /= n
accuracy
```
